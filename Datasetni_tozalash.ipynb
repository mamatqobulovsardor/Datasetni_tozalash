#Datasetni tozalash

```python
#Kutubxonalarni yuklash
import numpy as np
import pandas as pd
```

```python
#malumotlarni yuklash
data = pd.read_csv("/content/Original_Dataset.csv")
data
```

##Age(yosh) bulimini tuldiramiz

```python
#Salbiy yoki yo'q qiymatdagi yoshlarni to'g'irlaymiz
# Agar "Age" ustunida salbiy qiymat bo'lsa, uni musbatga o'zgartiramiz.
# Agar "Age" qiymati yo'q bo'lsa, uni o'rtacha yosh bilan to'ldiramiz (bu holda 30).
data['Age'] = data['Age'].apply(lambda x:abs(x)if x and x < 0 else(30 if pd.isnull(x) else x))
```

```python
data
```

yoshdaga qiymatlar uzgardi

##Name(ism) bulimidagi yuq qiymatlarga ism quyamis yo bulmasa Unknown yozamiz

```python
data['Name'] = data['Name'].fillna("Samir")
```

##Gender

```python
#"Gender"dagi bo'sh joylarni "Not Specified" bilan to'ldiramiz
data['Gender'] = data['Gender'].fillna("Not Specified")
```

##Salary(Maosh)

```python
data['Salary'] = data['Salary'].replace("Not Defined", np.nan).astype(float)
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
```

```python
data
```

##Join_Data
(sana qiymatlarini tug'irlaymiz)

```python
data['Join_Date'] = data['Join_Date'].fillna("2023-01-01")
data['Join_Date'] = pd.to_datetime(data['Join_Date'])
```
